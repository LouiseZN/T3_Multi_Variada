## Primeira questao

#### Hotellings test for multivariate mean in two independent samples

#### The null hypothesis is that both samples have the same mean. The alternative hypothesis is that means differ.

```python
#a)
df_a_mas = df_mas[["HRT1", "HRT2", "HRT3", "HRT4"]]
df_a_fem = df_fem[["HRT1", "HRT2", "HRT3", "HRT4"]]
test_mvmean_2indep(df_a_mas, df_a_fem)
```

```python
#b)
df_b_mas = df_mas[["COT1", "COT2", "COT3", "COT4"]]
df_b_fem = df_fem[["COT1", "COT2", "COT3", "COT4"]]
test_mvmean_2indep(df_b_mas, df_b_fem)
```

```python
# c) Rejeita H0 
df_c_mas = df_mas[["SVRIT1", "SVRIT2", "SVRIT3", "SVRIT4"]]
df_c_fem = df_fem[["SVRIT1", "SVRIT2", "SVRIT3", "SVRIT4"]]
test_mvmean_2indep(df_c_mas, df_c_fem)
```

```python
# d) Rejeita H0 
df_d_mas = df_mas[["PWVT1", "PWVT2", "PWVT3", "PWVT4"]]
df_d_fem = df_fem[["PWVT1", "PWVT2", "PWVT3", "PWVT4"]]
test_mvmean_2indep(df_d_mas, df_d_fem)
```
