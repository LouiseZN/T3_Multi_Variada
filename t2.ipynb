```python
# a)
import statsmodels.api as sm

variaveis = [col for col in df.columns if col not in ['Y1', 'Y2']]

X = df[variaveis]

#X.drop(["X6","X8"], axis=1, inplace=True)

X = sm.add_constant(X)

y = df['Y2']

# Ajusta OLS
fit0 = sm.OLS(y, X).fit()

print(fit0.summary())
```

```python
# b)
def stepwise_selection(X, y, threshold_out=0.05):
    initial_features = X.columns.tolist()
    best_features = initial_features.copy()
    
    while len(best_features) > 0:
        # Adiciona constante para o statsmodels funcionar direito
        X_const = sm.add_constant(X[best_features])
        model = sm.OLS(y, X_const).fit()
        
        # Pega o maior p-valor (ignorando a constante)
        p_values = model.pvalues.drop('const')
        max_p_value = p_values.max()
        feature_to_remove = p_values.idxmax()
        
        # Se o pior p-valor for maior que 0.05, remove a variável
        if max_p_value > threshold_out:
            print(f"Removendo '{feature_to_remove}' (p-valor: {max_p_value:.4f})")
            best_features.remove(feature_to_remove)
        else:
            break # Sai do loop se todas forem boas
            
    return best_features

# --- RODANDO O STEPWISE ---
print("--- Iniciando Stepwise ---")
variaveis_finais = stepwise_selection(X, y)

print(f"\nVariáveis que sobreviveram: {variaveis_finais}")

# Ajustando o modelo final com apenas as variáveis vencedoras
X_final = sm.add_constant(X[variaveis_finais])
modelo_final = sm.OLS(y, X_final).fit()

print(modelo_final.summary())
```

Mesmo revomevendo as duas colunas o R-squared permanece igual, e o AIC tem uma diferença de apenas 3 (3975-3972)

```python
# c)
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error

# 2. Divisão Treino-Teste (80-20) com semente fixa
# random_state=42 é a semente padrão para garantir que sempre dê o mesmo resultado
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)


# --- MODELO 1: COMPLETO (Full) ---
X_train_full = sm.add_constant(X_train)
X_test_full = sm.add_constant(X_test)

model_full = sm.OLS(y_train, X_train_full).fit()
y_pred_full = model_full.predict(X_test_full)

# --- MODELO 2: STEPWISE ---
# A seleção é feita olhando SÓ para o treino
vars_stepwise = stepwise_selection(X_train, y_train)
print(f"Variáveis selecionadas no treino: {vars_stepwise}")

X_train_step = sm.add_constant(X_train[vars_stepwise])
X_test_step = sm.add_constant(X_test[vars_stepwise]) # Aplica as mesmas vars no teste

model_step = sm.OLS(y_train, X_train_step).fit()
y_pred_step = model_step.predict(X_test_step)

# --- AVALIAÇÃO (Métricas no Teste) ---
mse_full = mean_squared_error(y_test, y_pred_full)
mae_full = mean_absolute_error(y_test, y_pred_full)

mse_step = mean_squared_error(y_test, y_pred_step)
mae_step = mean_absolute_error(y_test, y_pred_step)

print("\n--- COMPARAÇÃO NO TESTE ---")
print(f"Modelo Completo | MSE: {mse_full:.4f} | MAE: {mae_full:.4f}")
print(f"Modelo Stepwise | MSE: {mse_step:.4f} | MAE: {mae_step:.4f}")
```

Ambos os modelos possuem (em termos de MSE e MAE) valores muito proximos, o que indica que de fato as variaveis X6 e X8 não eram muito relevantes

```python
# c)
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Função auxiliar para calcular VIF e retornar um DataFrame bonitinho
def calcular_vif(X):
    # O VIF precisa da constante para ser calculado corretamente se houver intercepto
    if 'const' not in X.columns:
        X = sm.add_constant(X)
    
    vif_data = pd.DataFrame()
    vif_data["Variavel"] = X.columns
    vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    
    return vif_data.sort_values(by="VIF", ascending=False)

# --- APLICANDO AOS MODELOS ---

# 1. VIF do Modelo Completo (Todas as variáveis)
# X_train_full foi definido no exercício anterior
print("--- VIF: Modelo Completo ---")
print(calcular_vif(X_train_full))

# 2. VIF do Modelo Stepwise (Apenas as selecionadas)
# X_train_step foi definido no exercício anterior
print("\n--- VIF: Modelo Stepwise ---")
print(calcular_vif(X_train_step))
```

X2, X3 e X4 possuem um VIF infinito pela função, isso significa que eles possuem uma colinearidade perfeita (ou quase efeita), olhando o significado deles, podemos ver que de fato X2 é uma combinação linear perfeita de X3 e X4 
X2: Superfície
X3: Área da parede
X4: Área do telhado

X2 = X3 + 2*X4

X1 e X5 tambem possue um alto VIF, então vamos testar sem o X1

```python
X_final = X_train_step.drop(["X1", "X2"], axis=1)
print(calcular_vif(X_final))
## X4 e X5 ainda possuem alto VIF, possuem algum grau de colinearidade
```

```python
# e)
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# 1. Preparar os dados (Excluindo Y1 e Y2)
cols_X = [col for col in df.columns if col not in ['Y1', 'Y2']]
X = df[cols_X]

# 2. PADRONIZAÇÃO (Passo Obrigatório no PCA)
# O PCA é sensível à escala. Se não padronizar, a variável com maiores números domina.
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# 3. Aplicar PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# 4. Análise da Variância Explicada (Scree Plot)
plt.figure(figsize=(10, 5))
plt.bar(
    range(1, len(pca.explained_variance_ratio_) + 1), 
    pca.explained_variance_ratio_ * 100, 
    alpha=0.7, align='center', label='Variância Individual'
)
plt.step(
    range(1, len(pca.explained_variance_ratio_) + 1), 
    np.cumsum(pca.explained_variance_ratio_) * 100, 
    where='mid', color='red', label='Variância Acumulada'
)
plt.ylabel('Porcentagem de Variância Explicada')
plt.xlabel('Componentes Principais')
plt.title('Scree Plot - Variância por Componente')
plt.legend(loc='best')
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

# 5. Identificar a Variável Mais Relevante (Loadings do PC1)
loadings = pd.DataFrame(
    pca.components_.T, 
    columns=[f'PC{i+1}' for i in range(X.shape[1])], 
    index=X.columns
)

# Ordenar pelo valor absoluto do peso no PC1
loadings['Abs_PC1'] = loadings['PC1'].abs()
relevancia = loadings.sort_values(by='Abs_PC1', ascending=False)

print("--- Cargas (Pesos) das Variáveis no PC1 ---")
print(relevancia[['PC1']])

mais_relevante = relevancia.index[0]
peso = relevancia.iloc[0, 0]
print(f"\n>>> A variável mais relevante é '{mais_relevante}' com carga de {peso:.4f} no PC1.")
```

```python
df_completo = pd.read_csv("Chemical Composion of Ceramic.csv")
df_completo
```

```python
#a)

from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans

df = df_completo.drop(["Ceramic Name", "Part"], axis=1)

X = df

scores = []
k_values = range(2, 11) # Começa em 2, pois silhouette não existe para k=1

for i in k_values:
    # n_init='auto' remove avisos chatos em versões novas
    modelo = KMeans(n_clusters=i, random_state=42, n_init='auto')
    
    modelo.fit(X)
    
    # Calcula o score
    score = silhouette_score(X, modelo.labels_)
    scores.append(score)

# Plotando Corretamente
plt.figure(figsize=(8, 5))
plt.plot(k_values, scores, marker='o', linestyle='-') # K no eixo X, Score no Y
plt.title('Silhouette Score por número de clusters')
plt.xlabel('Número de Clusters (k)')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()

# Mostrando qual o melhor k
melhor_k = k_values[np.argmax(scores)]
print(f"O melhor número de clusters baseado no Silhouette é: {melhor_k}")
```
