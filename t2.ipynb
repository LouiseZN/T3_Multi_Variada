```python
from statsmodels.stats.multivariate import test_mvmean_2indep
import pandas as pd 
```

```python
df = pd.read_csv("laseri.csv")
df 
```

```python
df_a = df[[ "Sex","HRT1", "HRT2", "HRT3", "HRT4"]]
df_a
```

```python
df_fem = df_a[df["Sex"] == "Female"]
df_fem = df_fem.drop(columns="Sex")
df_fem
```

```python
df_mas = df_a[df["Sex"] == "Male"]
df_mas = df_mas.drop(columns="Sex")
df_mas
```

#### Hotellings test for multivariate mean in two independent samples

#### The null hypothesis is that both samples have the same mean. The alternative hypothesis is that means differ.

```python
test_mvmean_2indep(df_mas, df_fem)
```
